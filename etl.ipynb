```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
spell_list = "Sources\Spells.csv"
movie_1 = "Sources\Harry Potter 1.csv"
movie_2 = "Sources\Harry Potter 2.csv"
movie_3 = "Sources\Harry Potter 3.csv"

spells_df = pd.DataFrame(pd.read_csv(spell_list,sep=";"))
script_1_df = pd.DataFrame(pd.read_csv(movie_1,sep=";"))
script_2_df = pd.DataFrame(pd.read_csv(movie_2,sep=";"))
script_3_df = pd.DataFrame(pd.read_csv(movie_3,sep=";")).rename(columns=str.title)
```

```python
script_full = pd.concat([script_1_df,script_2_df,script_3_df])\
```

```python
spell_names_df = spells_df[spells_df["Incantation"] != "Unknown"].dropna()
spell_names_df = spell_names_df[spell_names_df["Incantation"] != "None"]
```

```python
spell_names = spell_names_df["Incantation"].tolist()
```

```python
wizard = []
says = []
spell = []

for index, row in script_full.iterrows():
    for spell_name in spell_names:
        if spell_name in row["Sentence"]:
            wizard.append(row["Character"])
            says.append(row["Sentence"])
            spell.append(spell_name)
            break
```

```python
filtered_dict = list(zip(wizard,says,spell))
filtered_df = pd.DataFrame(filtered_dict)
```

```python
final_df = filtered_df.rename(columns={0:"Wizard",1:"Sentence",2:"Spell"})
final_df["Wizard"] = final_df["Wizard"].str.title()
```
